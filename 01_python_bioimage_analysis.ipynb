## Bioio

Bioio is a python created to offer support on most commercial and open source microcopy file format.

Here we will use the downloaded dataset , 🖼️ : `hela_prolongdiamond_dapi_tubulin-af488_mitochondria-af568_40x.czi` A.K.A. `image.czi`, a czi image with 4 scenes, 3 channels, 14 Z, 512x512px

### Import the library

````python
from bioio import BioImage
````

### Read the scenes information

````python
image_path = "/content/hela_prolongdiamond_dapi_tubulin-af488_mitochondria-af568_40x.czi"

# Get a BioImage object
img = BioImage(image_path)  # selects the first scene found

# Get the id of the current operating scene
print(img.current_scene)

# Get a list valid scene ids
print(img.scenes)
````

### Access to a specific scene

By default, once you do `img = BioImage(image_path)`, you access to the first scene. If you want to open another one do

````python
img = BioImage(image_path)

# Change scene using name
img.set_scene("1")
# Or by scene index
img.set_scene(1)
````

### Get the image information

Get the image information such as dimension, axis order etc...

````python
print(img.dims)  # returns a Dimensions object
print(img.dims.order)  # returns string "TCZYX"
print(img.dims.X)  # returns size of X dimension
print(img.shape)  # returns tuple of dimension sizes in TCZYX order
````

### Get the image data as a numpy array

````python
data=img.get_image_data("CZYX", T=0)  # returns 4D CZYX numpy array
print(data.shape)
print(type(data))  # returns <class 'numpy.ndarray'>
print(data.dtype)
````

### Get the Metadata

Get the metadata such as pixel size, channel names etc...

````python
from bioio import BioImage

# Get a BioImage object
img = BioImage(image_path)  # selects the first scene found
img.metadata  # returns the metadata object for this file format (XML, JSON, etc.)
print(img.channel_names)  # returns a list of string channel names found in the metadata
print(img.physical_pixel_sizes.Z)  # returns the Z dimension pixel size as found in the metadata
print(img.physical_pixel_sizes.Y)  # returns the Y dimension pixel size as found in the metadata
print(img.physical_pixel_sizes.X)  # returns the X dimension pixel size as found in the metadata
print(img.ome_metadata)
````

### Write the ome.tiff

````python
from bioio_ome_tiff.writers import OmeTiffWriter
import bioio_base as biob



output_path="/content/output.ome.tiff"

data=img.get_image_data("CZYX", T=0)  # returns 4D CZYX numpy array


write_dim_order = "CZYX"  # required, the dimension order to write the data in
channel_names = ['Mitochondria', 'Tubulin', 'DAPI']
channel_colors = [(255,0,255), (0,255,0), (0,255,255)]  # Magenta/Green/Cyam, colors in RGB format
pixel_size = biob.types.PhysicalPixelSizes(img.physical_pixel_sizes.Z, img.physical_pixel_sizes.Y, img.physical_pixel_sizes.X)  # optional, Z, Y, X in microns


OmeTiffWriter.save(
        data,
        output_path,
        write_dim_order,
        channel_names=channel_names,
        channel_colors=channel_colors,
        physical_pixel_sizes=pixel_size,
    )

````

## Import

````python
from cellpose import models
import numpy as np
import matplotlib.pyplot as plt
from cellpose import plot
````

## Configuration

````python
# Get a BioImage object
img = BioImage(image_path)  # selects the first scene found

print("img shape: {}".format(img.shape))
print("Dimension order: {}".format(img.dims.order))

image_scale = [img.physical_pixel_sizes.Z, img.physical_pixel_sizes.Y, img.physical_pixel_sizes.X]
print("Dimension order: {}".format(image_scale))

color_map = 'cyan'

nuclei_cell_image = img.get_image_data("CZYX", T=0)[[1, 2], :, :, :].squeeze()
````

### Import the CellPose model

<font color='red'>Carefull since the last CellPose model (cpsam) is 1.15Gb</font

````python
model = models.CellposeModel(gpu=True)
````

### Run the Model in 2D

#### Maximum Intensity projection along the Z axis

````python
print("nuclei_cell_image shape: {}".format(nuclei_cell_image.shape))
# Create from the Z stack a Maximum Intensity projection through the axis 1 which in that case the Z axis in the numpy array
nuclei_cell_image_MIP= np.max(nuclei_cell_image, axis=1)
print("nuclei_cell_image MIP shape: {}".format(nuclei_cell_image_MIP.shape))
````

````python
from cellpose import models

# Initialize model with GPU support based on argument
#model = models.CellposeModel(gpu=True)

#define the parameters
diameter = 60 # in pixels, set to 0 to let the model estimate the diameter
channel_axis = 0 # first axis is the channel axis
do_3D = False # enable 3D processing
cellprob_threshold = 0.0 # all pixels with value above threshold kept for masks, decrease to find more and larger masks. Defaults to 0.0.
flow_threshold = 0.4 #if do_3D and flow3D_smooth>0, smooth flows with gaussian filter of this stddev. Defaults to 0.

# Run segmentation with 2D
masks, flows, styles = model.eval(
    nuclei_cell_image_MIP,
    diameter=diameter,
    channel_axis=channel_axis,
    do_3D=do_3D,
    cellprob_threshold=cellprob_threshold,
    flow_threshold=flow_threshold,
)
````

#### Plot the result of the segmentation

````python
import matplotlib.pyplot as plt
from cellpose import plot

channels = [0,1]
fig = plt.figure(figsize=(12,5))
plot.show_segmentation(fig, nuclei_cell_image_MIP.squeeze(), masks, flows[0], channels=channels)
plt.tight_layout()
plt.show()
````

#### Display with stackview

Labels will be colored if the numpy array is of dtype int32

````python
stackview.curtain( nuclei_cell_image_MIP[0].squeeze(), masks.astype(np.int32))
````

### Run the Model in 3D

#### Get the anisotropy

````python
anisotropy = img.physical_pixel_sizes.Z / img.physical_pixel_sizes.X
print("Anisotropy (Z/X): {:.2f}".format(anisotropy))
````

````python
#define the parameters
diameter = 60 # in pixels, set to 0 to let the model estimate the diameter
channel_axis = 0 # first axis is the channel axis
z_axis = 1 # second axis is the z axis
do_3D = True # enable 3D processing
anisotropy = anisotropy # anisotropy of the dataset
cellprob_threshold = 0.0 # all pixels with value above threshold kept for masks, decrease to find more and larger masks. Defaults to 0.0.
flow_threshold = 0.4 #if do_3D and flow3D_smooth>0, smooth flows with gaussian filter of this stddev. Defaults to 0.
min_size = 2000 #all ROIs below this size, in pixels, will be discarded. Defaults to 15.
flow3D_smooth = 1 #if do_3D and flow3D_smooth>0, smooth flows with gaussian filter of this stddev. Defaults to 0.

# Run segmentation with 3D processing if enabled
masks, flows, styles = model.eval(
    nuclei_cell_image,
    diameter=diameter,
    channel_axis=channel_axis,
    z_axis=z_axis,
    do_3D=do_3D,
    anisotropy=anisotropy,
    flow3D_smooth=flow3D_smooth,
    cellprob_threshold=cellprob_threshold,
    flow_threshold=flow_threshold,
    min_size=min_size
)
````

# Plot Result

## Import

````python
import pandas as pd
import numpy as np  # the go-to package to deal with array data
import seaborn as sns  # a more advanced and automatized plotting library
from matplotlib import pyplot as plt  # the go-to package to create visualization
from skimage import measure
````

````python
feature_list = feature_list=[
        'label',
        'centroid',
        'area',
        'max_intensity',
        'mean_intensity',
        'min_intensity'
    ]

result_dataframe = pd.DataFrame(
                measure.regionprops_table(
                    masks,
                    intensity_image=nuclei_cell_image[0].squeeze(),
                    properties=feature_list,
                )
            )
````

````python
result_dataframe.head(20)
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(data=result_dataframe, x="mean_intensity", kde=True, color="skyblue")
plt.title("Distribution of mean intensity per object")
plt.show()
````

````python
sns.boxplot(data=result_dataframe[["max_intensity","mean_intensity","min_intensity"]])
plt.title("Distribution of intensity features")
plt.ylabel("Intensity")
plt.show()
````

````python
sns.pairplot(result_dataframe[["area", "mean_intensity", "max_intensity", "min_intensity"]])
plt.show()
````

````python
result_dataframe["centroid_y"] = result_dataframe["centroid-1"]
result_dataframe["centroid_x"] = result_dataframe["centroid-2"]

sns.scatterplot(data=result_dataframe, x="centroid_x", y="centroid_y", size="area", hue="mean_intensity", palette="viridis")
plt.gca().invert_yaxis()  # to match image coordinates
plt.title("Object positions with size & intensity")
plt.show()
````

## Seaborn Reference:

````python
# Create a figure with two panels
f, axs = plt.subplots(1, 2, figsize=(10, 4))

# first pannel, plot the area over time per positon
feature = "area"
sns.boxplot(
    x="label",
    y=feature,
    data=result_dataframe,
    ax=axs[0],
)
axs[0].set_title("Volume of cell")

# second pannel: nucleus eccentricity (overall)
feature = "mean_intensity"
sns.boxplot(
    x="label",
    y=feature,
    data=result_dataframe,
    ax=axs[1],
    fliersize=0,  # <- don't show outliers
)
axs[1].set_title("Overall nuclei mean intensity")

# remove top and right axes spines to be prettier
for ax in axs:
    sns.despine(ax=ax)
````
